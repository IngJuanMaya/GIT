## 3) If garbage data-in then garbage data-out.

Ejemplos:

* Diagnostico covid erroneo. Causa, "El uso de datos mal etiquetados o datos de fuentes desconocidas fue un culpable común".

* Compra y venta de propiedades con estimación de valor y predicción de posibilidad de venta erroneo. Causa, especulación de la precisión de la herramienta causo merma en los inversionistas interesados -> variable: percepción del mercado sobre la confiabilidad de la herramienta.

* Racismo en herramientas de clasificación medica para beneficios de atención. Causa "Es más probable que las personas de color tengan ingresos más bajos, lo que, incluso cuando están asegurados, puede hacer que tengan menos probabilidades de acceder a la atención médica. El sesgo implícito también puede hacer que las personas de color reciban una atención de menor calidad".

* Un chatbot de Microsoft que publicó tuits racistas. Causa, "...Microsoft no tuvo en cuenta fue que un grupo de usuarios de Twitter inmediatamente comenzaría a publicar en esa red social comentarios racistas y misóginos a Tay. El bot aprendió rápidamente de ese material y lo incorporó a sus propios tweets".

* Una herramienta de reclutamiento de Amazon que solo recomendaba hombres. Causa, "se entrenaron con currículos de 10 años enviados a Amazon, la mayoría de ellos de hombres".

otro tipo de errores:

* Falta de objetivos claros.

* Poca experticie.

* Baja aceptación por no gestionar el cambio.

* Baja calidad de los datos.

* No involucrar a las partes interesadas.

* No entender los resultados (abundancia de modelos black box).

* Testeo y validación inadecuados. Sesgos de confirmación.

* Falta de planeación a largo plazo.

* Mantenimiento de los modelos. Absolecencia.

* omitir o desconocer las consderaciones eticas y legales exigidas en el pais-sector-localidad de despliegue.

* Usuario y desarrolladores descoordinados - no se hablan entre si las espectativas.
